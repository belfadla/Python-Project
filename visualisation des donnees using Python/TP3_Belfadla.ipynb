```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
from pandas.plotting import scatter_matrix
```

```python
import seaborn as sns
```

```python
data = pd.read_csv("caffeine.csv",encoding='utf-8')
data.head()
```

```python
len(data)
```

```python
df = data.dropna()
df
```

```python
len(df)
```

```python
#Caffeine (mg)VS Volume (ml)
plt.scatter( df['Caffeine'], df['Volume'], marker='x',
            color='b', alpha=0.7, s = 124)
plt.title("volume vs caffeine")
plt.xlabel("Caffeine (mg)")
plt.ylabel("Volume (ml)")
plt.show()
```

```python
scatter_matrix(df)
plt.show()
```
